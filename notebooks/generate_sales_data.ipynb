```python
import random
import numpy as np
import pandas as pd


transaction_dates = pd.date_range(
    start="2024-01-01",
    periods=365,
    freq="D"
)
products = [
    'Protein Powder',
    'Creatine Monohydrate'
]
flavours = [
    'Cookies and Cream'
    'Strawberry',
    'Apple Pie',
    'Chocolate',
    'Vanilla'
]

# Simulate sales data
data = []
for date in transaction_dates:
    for product in products:
        sales = np.random.randint(20, 100)  # Random units sold

        if product == "Protein Powder":
            price = 39.99
            flavour = random.choice(flavours)
        else:
            price = 29.99
            flavour = 'Unflavoured'
         
        data.append([date, product, flavour, sales, price])

sales_df = pd.DataFrame(
    data,
    columns=['date','product','flavour','units_sold','price']
)
sales_df.to_csv(
    '../data/sales_data.csv',
    index=False
)
```
